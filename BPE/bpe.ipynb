```python
tokens = list(text.encode("utf-8")) # raw bytes of text --> int
print(len(tokens))
```

```python
def pairing(tokens): # pair baisis of frequency
    pairs = {}
    for pair in zip(tokens,tokens[1:]):
        pairs[pair] = pairs.get(pair,0)+1
    return pairs
```

```python
pairs = pairing(tokens)
top_pair = max(pairs,key=pairs.get)
```

```python
def merge(old_tokens,top_pair,idx):
    new_tokens = []
    i=0
    while i<len(old_tokens):
        if i<len(old_tokens)-1 and top_pair[0] ==  old_tokens[i] and top_pair[1] == old_tokens[i+1]:
            new_tokens.append(idx)
            i+=2
        else:
            new_tokens.append(old_tokens[i])
            i+=1
    return new_tokens
```

```python
new_tokens = merge(tokens,top_pair,256)
len(new_tokens)
```

```python
merges = {}

merges[top_pair] = 256
vocab_size  = 449

num_merges = vocab_size-256
org_tokens = tokens

for i in range(num_merges):
    pair = pairing(new_tokens)
    top_pair = max(pair,key=pair.get)
    idx = 257+i
    new_tokens = merge(new_tokens,top_pair,idx)
    merges[top_pair] = idx

```

```python
len(new_tokens)
```

```python
print(f" size compressed by {len(org_tokens)/len(new_tokens):.3f}X")
```

```python
vocab = {idx:bytes([idx]) for idx in range(256)}

for (p0,p1),v in merges.items():
    vocab[v] = vocab[p0] + vocab[p1]
```

```python
def decode(ids): # ids to text
    tokens = b"".join(vocab[idx] for idx in ids)
    text = tokens.decode('utf-8',errors='replace')
    return text
```

```python
decode([100,101,102,103])
```

```python
def encode(text):
    tokens = text.encode("utf-8")

    while 2<=len(tokens):
        pairs = pairing(tokens)
        bottom_pair = min(pairs,key = lambda p: pairs.get(p,float("inf")))

        if bottom_pair not in merges:
            break

        idx = merges[bottom_pair]
        tokens = merge(tokens,bottom_pair,idx)
    return tokens
```
